```python
import re
```

```python
# открывем файл
with open('пример_блокир.txt', 'r') as file:
    text = file.read()

# Используем регулярные выражения для поиска слов, содержащих небуквенные символы
search = re.compile(r'\b\w*[^a-zA-Z ]\w*\b') # слова, которые начинаются с нуля или более букв или цифр, за которыми следует любой символ, кроме буквы или пробела, за которым следует ноль или более букв или цифр.
words = search.findall(text)
# слова, которые написаны правильно, но содержат специальные символы
misspelled_words = []
for match in words: # проходим по каждому слову из списка words
    if not re.match(r'^\w+$', match): # смотрим на соответствие ^\w+$ - строка, которая состоит только из буквенно-цифровых символов без пробелов иди др. символов
        misspelled_words.append(match)
#печать
print("Misspelled words: ", misspelled_words)
```

```python
# Загрузим словарь
d = enchant.Dict("en_US")

# откроем текст
with open("пример_блокир.txt", "r", encoding='utf-8') as open_file: 
    text = open_file.read()
    
words = text.split()# разделим 

# Проверка слов
misspelled_words = []
for word in words:
    if not d.check(word): #check проверяет слова
        misspelled_words.append(word)

# вывод слов
print("Misspelled words: ", misspelled_words)
```

```python
#видим, что вывадся & и ... 
text = re.sub(r'[&…]', '', text) #Sub возвращает новую строку, полученную в результате замены по указанному шаблону
```

```python
# вывод слов с частотностью 
misspelled_words = {}
for word in words:
    if not d.check(word): # check проверяет слово 
        if word in misspelled_words:
            misspelled_words[word] += 1
        else:
            misspelled_words[word] = 1

# вывод + частотность ошибок
print("Misspelled words: ")
for word, frequency in misspelled_words.items():
    print(word, ":", frequency)
```

```python
# поробуем исправить слова 
for word in words:
  if not d.check(word):
    suggestions = d.suggest(word) # Мы можем вывести список возможных исправлений слова c d.suggest
    print (f"Misspeled word: {word}. Suggestions:{suggestions}")
```

```python
# мы можем увидеть, что 0 можно заменить на о, @ на a, 3 на e
```

```python
# заменим их на буквы
patterns = {
  
    "@": "a",
    "0": "o",
    "!": "i",
    "3": "e",
    "IV3RM3CTIN": "ivermectin"
}
for pattern, replacement in patterns.items():
    text = re.sub(pattern, replacement, text)

print(text)
```
